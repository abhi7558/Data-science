```python
print("Numeric data types and operations\n") 
a = 9
b = 12
print("a = ",a,"b = ",b)
print("\n")

print("Arithmetic operations\n")
sum = a + b
difference = a - b
product = a * b
division = a / b


print(" a + b =",sum)
print(" a - b = ",difference)
print(" a * b =",product)
print(" a / b = ",division)

print("\n")
print("Boolean Operations\n")
print("a is greater than b :",a>b)
print("a is less than b :",a<b)
print("a not equal to b :",a!=b)
print("a  equal to b :",a==b)
print("\n")




print("Updating Values")
print("\n")
# Updating values
a += 1
b *= 2
print("After updating: a = ",a," b = ",b)
print("\n")
print("\n")



# String operations
print("String operations\n")
string1 = "hello"
string2 = "world"
print("String1:", string1," String2: ",string2)
print("Length of string1: ",len(string1))
print("Length of string2: ",len(string2))
concat = string1 + " " + string2
print("Concatenated string: ",concat)
print("Uppercase string1: ",string1.upper())
print("Capitalized string2: ",string2.capitalize())
print("Centered string1: ",string1.center(10))
print("Justified string2: ",string2.ljust(10))
print("String2 replace 'o' with 'e': ",string2.replace('o', 'e'))
print(f"Trimmed string with whitespace: ",string1)
formatted_string = f"Value of a: {a}, Value of b: {b}"
print("Formatted string: ",formatted_string)
print("\n")
print("\n")

print("Dictionary\n")
dict = {
    1:"Abhi",
    2:"Sinan",
    3:"Yadu",
    }
print("Dictionary",dict)

# Length of the dictionary
print("Number of items in the dictionary:", len(dict))





```
